```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(2026)

n = 150
```

```python
# Lista de bairros usada como base para o notebook
bairros = [
    "Efapi", "Centro", "Jardim América", "Passo dos Fortes",
    "Palmital", "São Cristóvão", "Universitário",
    "Presidente Médici", "Santa Maria", "Vila Real"
]
```

# 1. Quem é o gerador?

```python
"""
tipo_gerador

Pessoa física (residência) — 55%
Domicílios representam a maior parcela de geradores em municípios brasileiros,
sendo responsáveis por cerca de 50-60% do volume total de resíduos sólidos.
Fonte: Renda e evolução da geração per capita de resíduos sólidos no Brasil —
SciELO, 2012.
https://www.scielo.br/j/esa/a/kZn74jmyqBL5GNT4yxkD8Jk/

Microempreendedor individual (MEI) — 20%
Pequenos comércios e prestadores de serviço formais são numerosos em cidades
de médio porte como Chapecó, gerando resíduos de forma irregular e pouco
monitorada.
Fonte: SNIS — Série Histórica, dados por município.
https://dados.gov.br/dados/conjuntos-dados/snis---srie-histrica

Pequena empresa — 15%
Empresas de pequeno porte contribuem com parcela significativa de resíduos
industriais leves, de alimentos e comerciais em municípios do Oeste de SC.
Fonte: SNIS — Série Histórica.

Instituição pública — 10%
Escolas, postos de saúde e órgãos municipais representam uma fração menor,
porém relevante pela regularidade e volume previsível de geração.
Fonte: MUNIC 2023 — IBGE.
https://agenciadenoticias.ibge.gov.br/agencia-noticias/2012-agencia-de-noticias/noticias/41994
"""

opcoes_tipo_gerador = [
    "Pessoa física (residência)",
    "Microempreendedor individual (MEI)",
    "Pequena empresa",
    "Instituição pública"
]

probs_tipo_gerador = [0.55, 0.20, 0.15, 0.10]

tipo_gerador = np.random.choice(
    opcoes_tipo_gerador,
    size=n,
    p=probs_tipo_gerador
)

tipo_gerador[:5]
```

```python
"""
tipo_atividade

Residencial — 55%
Corresponde diretamente à proporção de geradores do tipo pessoa física.

Alimentação (restaurante, padaria, mercado, etc.) — 12%
Estabelecimentos alimentícios são numerosos em centros urbanos e geram
grande volume de resíduos orgânicos.
Fonte: ABRELPE Panorama 2023 — geração por setor.
https://www.abrema.org.br/wp-content/uploads/dlm_uploads/2024/03/Panorama_2023_P1.pdf

Comércio — 13%
Comércios em geral (vestuário, eletrônicos, materiais) geram resíduos secos
como papelão, plástico e embalagens.

Serviços — 10%
Inclui escritórios, clínicas e prestadores de serviços formais.

Indústria de alimentos — 5%
Chapecó possui forte presença do agronegócio e processamento de alimentos,
como frigoríficos e cooperativas.
Fonte: Panorama socioeconômico de Chapecó — Prefeitura Municipal.

Indústria em geral — 5%
Indústrias diversas de pequeno porte com geração variada de resíduos.
"""

opcoes_tipo_atividade = [
    "Residencial",
    "Alimentação (restaurante, padaria, mercado, etc.)",
    "Comércio",
    "Serviços",
    "Indústria de alimentos",
    "Indústria em geral"
]

probs_tipo_atividade = [0.55, 0.12, 0.13, 0.10, 0.05, 0.05]

tipo_atividade = np.random.choice(
    opcoes_tipo_atividade,
    size=n,
    p=probs_tipo_atividade
)

tipo_atividade[:5]
```

```python
"""
Bairros gerados com distribuição uniforme entre as opções listadas.
Hipótese neutra, usada apenas até termos a frequência real dos bairros no formulário.
"""

bairro = np.random.choice(
    bairros,
    size=n
)

bairro[:5]
```

# 2. O que ele gera?

```python
"""
tipo_residuo_principal

Orgânicos — 45%
Resíduos orgânicos representam entre 45% e 55% da composição gravimétrica
dos resíduos domiciliares brasileiros, sendo a maior fração gerada.
Fonte: Composição gravimétrica e estimativa de geração per capita — RS, 2020.
http://revista.ecogestaobrasil.net/v7n16/v07n16a31.pdf

Plástico — 20%
Plásticos são a segunda maior fração em volume, com crescimento contínuo
associado ao consumo de embalagens descartáveis.
Fonte: Brasil é um dos principais geradores de resíduo plástico no mundo.
https://interacoes.ucdb.br/interacoes/article/download/3671/2784

Papelão — 15%
Papelão e papel representam fração expressiva, especialmente em comércios
e domicílios com consumo de e-commerce.

Vidro — 8%
Fração estável, gerada principalmente por bebidas e alimentos envasados.

Metal — 7%
Latas de alumínio e embalagens metálicas compõem parcela menor mas de
alto valor para reciclagem.

Óleo residual — 5%
Óleos de fritura são gerados em menor volume, mas com impacto ambiental
elevado quando descartados inadequadamente.
Fonte: ABRELPE Panorama 2023.
https://www.abrema.org.br/wp-content/uploads/dlm_uploads/2024/03/Panorama_2023_P1.pdf
"""

opcoes_tipo_residuo = [
    "Orgânicos",
    "Plástico",
    "Papelão",
    "Vidro",
    "Metal",
    "Óleo residual"
]

probs_tipo_residuo = [0.45, 0.20, 0.15, 0.08, 0.07, 0.05]

tipo_residuo_principal = np.random.choice(
    opcoes_tipo_residuo,
    size=n,
    p=probs_tipo_residuo
)

tipo_residuo_principal[:5]
```

```python
"""
residuo_organico_principal

Restos de comida — 35%
Principal fração orgânica em domicílios brasileiros, representando sobras
de refeições do dia a dia.
Fonte: Estudo do perfil de resíduos sólidos domésticos — REMA, 2021.
https://editoraime.com.br/revistas/index.php/rema/article/view/2730

Cascas de legumes e verduras — 18%
Segunda fração mais relevante, gerada no preparo de refeições.

Cascas de frutas — 15%
Presente em domicílios e estabelecimentos de alimentação.

Frutas estragadas — 12%
Geradas por desperdício alimentar, comum tanto em domicílios quanto em
comércios de alimentos.

Casca de ovos — 10%
Resíduo frequente em domicílios e restaurantes.

Borra de café — 10%
Gerada diariamente em domicílios e estabelecimentos comerciais.
"""

opcoes_residuo_organico = [
    "Restos de comida",
    "Cascas de legumes e verduras",
    "Cascas de frutas",
    "Frutas estragadas",
    "Casca de ovos",
    "Borra de café"
]

probs_residuo_organico = [0.35, 0.18, 0.15, 0.12, 0.10, 0.10]

residuo_organico_principal = np.random.choice(
    opcoes_residuo_organico,
    size=n,
    p=probs_residuo_organico
)

residuo_organico_principal[:5]
```

```python
"""
origem_residuo

Consumo diário — 60%
A maior parte dos resíduos domiciliares e de pequenos geradores é originada
do consumo cotidiano de alimentos, embalagens e produtos de higiene.
Fonte: Gestão de resíduos domiciliares — SciELO, 2023.
https://www.scielo.br/j/urbe/a/5YQCmrGs9Mh3s3LDpsQY8jD/

Processo produtivo — 20%
Atividades produtivas de pequenas indústrias e estabelecimentos alimentícios
geram resíduos como subprodutos do processamento.

Sobra de matéria-prima — 12%
Resíduos de matéria-prima não utilizada, comum em pequenas empresas
de manufatura e alimentação.

Produto fora de especificação — 8%
Produtos com defeito ou vencidos, descartados por estabelecimentos
comerciais e indústrias.
"""

opcoes_origem_residuo = [
    "Consumo diário",
    "Processo produtivo",
    "Sobra de matéria-prima",
    "Produto fora de especificação"
]

probs_origem_residuo = [0.60, 0.20, 0.12, 0.08]

origem_residuo = np.random.choice(
    opcoes_origem_residuo,
    size=n,
    p=probs_origem_residuo
)

origem_residuo[:5]
```

# 3. Quanto ele gera?

```python
"""
quantidade_semanal

No Brasil, a geração per capita domiciliar é de aproximadamente 1 kg/dia por
habitante, o que equivale a 7 kg/semana por pessoa. Para famílias de 3 a 4
pessoas, a faixa de 5 a 10 kg/semana é a mais representativa.
Fonte: Estudo do perfil de resíduos sólidos domésticos — REMA, 2021.
https://editoraime.com.br/revistas/index.php/rema/article/view/2730

1 a 5 kg por semana — 30%
Domicílios menores ou com práticas de redução de resíduos.

5 a 10 kg por semana — 35%
Faixa mais comum para domicílios de 3 a 4 pessoas.

10 a 50 kg por semana — 20%
Famílias maiores ou pequenos estabelecimentos comerciais.

50 a 200 kg por semana — 7%
Estabelecimentos de médio porte, como restaurantes e mercados.

Mais de 200 kg por semana — 3%
Grandes geradores, como indústrias ou estabelecimentos de grande porte.

Não sei estimar — 5%
Parcela de respondentes sem controle sobre a quantidade gerada.
"""

opcoes_quantidade = [
    "1 a 5 kg por semana",
    "5 a 10 kg por semana",
    "10 a 50 kg por semana",
    "50 a 200 kg por semana",
    "Mais de 200 kg por semana",
    "Não sei estimar"
]

probs_quantidade = [0.30, 0.35, 0.20, 0.07, 0.03, 0.05]

quantidade_semanal = np.random.choice(
    opcoes_quantidade,
    size=n,
    p=probs_quantidade
)

quantidade_semanal[:5]
```

```python
"""
num_pessoas

Baseado no perfil demográfico de Chapecó e médias nacionais do IBGE,
onde domicílios urbanos têm em média 2,7 a 3,2 pessoas por residência.
Fonte: IBGE Censo 2022 — Resultados preliminares.
https://agenciadenoticias.ibge.gov.br

1 a 3 pessoas — 35%
Domicílios menores, típicos de jovens e casais sem filhos.

4 a 6 pessoas — 35%
Família média brasileira; faixa mais representativa.

6 a 10 pessoas — 15%
Famílias maiores ou pequenos estabelecimentos com poucos funcionários.

10 a 30 pessoas — 10%
Pequenas empresas ou estabelecimentos comerciais de pequeno porte.

30 a 50 pessoas — 3%
Empresas de médio porte.

Mais de 50 pessoas — 2%
Grandes empresas ou instituições públicas.
"""

opcoes_num_pessoas = [
    "1 a 3",
    "4 a 6",
    "6 a 10",
    "10 a 30",
    "30 a 50",
    "Mais de 50 pessoas"
]

probs_num_pessoas = [0.35, 0.35, 0.15, 0.10, 0.03, 0.02]

num_pessoas = np.random.choice(
    opcoes_num_pessoas,
    size=n,
    p=probs_num_pessoas
)

num_pessoas[:5]
```

# 4. O que faz com o resíduo?

```python
"""
destino_atual

Coleta pública comum — 60%
No Brasil, cerca de 95% dos resíduos urbanos ainda são enviados a aterros ou
lixões, enquanto menos de 5% passa por reciclagem ou compostagem em nível
municipal. Estudos em municípios médios mostram que a maioria descarta no
caminhão comum, mesmo quando existe coleta seletiva disponível.
Fonte: Development of waste-to-energy through integrated sustainable waste
management — PMC, 2023.
https://pmc.ncbi.nlm.nih.gov/articles/PMC9838418/

Coleta seletiva — 20%
Pesquisas de percepção em cidades brasileiras indicam que uma parcela
minoritária, porém significativa, declara separar resíduos e utilizar coleta
seletiva quando o serviço existe, ainda que de forma irregular.
Fonte: Evaluation of a municipal program of selective collection — UEM, 2013.
https://periodicos.uem.br/ojs/index.php/ActaSciTechnol/article/download/16095/pdf

Compostagem — 7%
A compostagem ainda é pouco praticada em domicílios urbanos brasileiros.
Fonte: Economic analysis of solid waste gravimetric composition — 2022.
https://studiespublicacoes.com.br/ojs/index.php/seas/article/view/442

Reaproveitamento interno — 5%
Parte dos resíduos é reaproveitada no próprio local, especialmente em
contextos rurais ou produtivos.
Fonte: Pilot Study of Generation and Disposal of Municipal Solid Wastes — 2014.
https://www.sciendo.com/article/10.2478/pjct-2014-0030

Doação — 8%
A doação de materiais ocorre em escala baixa, mas presente em programas
locais e iniciativas sociais.
Fonte: Social innovation in waste management in residential buildings — 2023.
https://revistas.pucsp.br/index.php/risus/article/download/61988/43173
"""

opcoes_destino_atual = [
    "Coleta pública comum",
    "Coleta seletiva",
    "Compostagem",
    "Reaproveitamento interno",
    "Doação"
]

probs_destino_atual = [0.60, 0.20, 0.07, 0.05, 0.08]

destino_atual = np.random.choice(
    opcoes_destino_atual,
    size=n,
    p=probs_destino_atual
)

destino_atual[:5]
```

```python
"""
custo_destinacao

Não — 55%
A maioria dos domicílios e pequenos geradores não tem custo associado ao
descarte, pois utiliza coleta pública gratuita.

Sim — 25%
Empresas e geradores especiais frequentemente pagam por serviços de coleta
ou destinação específica (aterro industrial, empresa de logística reversa).

Não sei — 20%
Parcela de respondentes sem clareza sobre custos indiretos da destinação.
"""

opcoes_custo = ["Sim", "Não", "Não sei"]
probs_custo   = [0.25, 0.55, 0.20]

custo_destinacao = np.random.choice(
    opcoes_custo,
    size=n,
    p=probs_custo
)

custo_destinacao[:5]
```

# 5. Percepção e comportamento

```python
"""
potencial_reaproveitamento

Sim — 50%
Metade dos respondentes reconhece que seus resíduos poderiam ser reaproveitados,
o que indica abertura para educação ambiental e mudança de comportamento.
Fonte: Gestão de resíduos domiciliares — SciELO, 2023.
https://www.scielo.br/j/urbe/a/5YQCmrGs9Mh3s3LDpsQY8jD/

Nunca pensei sobre isso — 30%
Parcela significativa sem reflexão prévia sobre o tema; alvo prioritário
de campanhas de conscientização.

Não — 20%
Respondentes que acreditam que seus resíduos não têm potencial de reuso.
"""

opcoes_potencial = ["Sim", "Não", "Nunca pensei sobre isso"]
probs_potencial  = [0.50, 0.20, 0.30]

potencial_reaproveitamento = np.random.choice(
    opcoes_potencial,
    size=n,
    p=probs_potencial
)

potencial_reaproveitamento[:5]
```

```python
"""
caracteristica_residuo

Alto teor orgânico — 30%
Reflete a composição gravimétrica brasileira, onde orgânicos dominam.
Fonte: Composição gravimétrica RS — 2020.
http://revista.ecogestaobrasil.net/v7n16/v07n16a31.pdf

É úmido — 25%
Característica associada aos resíduos orgânicos e de alimentos.

Contém vitaminas, minerais e compostos nutritivos — 15%
Respondentes com maior conhecimento sobre potencial dos resíduos orgânicos.

É seco — 15%
Resíduos secos como papel, plástico e metal.

Não sei informar — 15%
Respondentes sem conhecimento técnico para classificar.
"""

opcoes_caracteristica = [
    "Alto teor orgânico",
    "É úmido",
    "Contém vitaminas, minerais e compostos nutritivos",
    "É seco",
    "Não sei informar"
]

probs_caracteristica = [0.30, 0.25, 0.15, 0.15, 0.15]

caracteristica_residuo = np.random.choice(
    opcoes_caracteristica,
    size=n,
    p=probs_caracteristica
)

caracteristica_residuo[:5]
```

```python
"""
tentativa_reutilizacao

Não — 70%
A maioria dos geradores nunca tentou reutilizar seus resíduos, o que é
consistente com a baixa taxa de compostagem e reaproveitamento no Brasil.

Sim — 30%
Parcela que já fez alguma tentativa de reutilização (compostagem caseira,
doação, reuso de embalagens, etc.).
"""

opcoes_tentativa = ["Sim", "Não"]
probs_tentativa  = [0.30, 0.70]

tentativa_reutilizacao = np.random.choice(
    opcoes_tentativa,
    size=n,
    p=probs_tentativa
)

tentativa_reutilizacao[:5]
```

```python
"""
interesse_aprender

Sim — 55%
Mais da metade dos respondentes demonstra interesse em aprender práticas
de reaproveitamento, indicando potencial para programas educativos.
Fonte: Social innovation in waste management — PUC-SP, 2023.
https://revistas.pucsp.br/index.php/risus/article/download/61988/43173

Talvez — 30%
Respondentes indecisos, mas receptivos se houver facilidade de acesso.

Não — 15%
Parcela sem interesse em mudança de comportamento.
"""

opcoes_interesse = ["Sim", "Talvez", "Não"]
probs_interesse  = [0.55, 0.30, 0.15]

interesse_aprender = np.random.choice(
    opcoes_interesse,
    size=n,
    p=probs_interesse
)

interesse_aprender[:5]
```

# 6. Montar o DataFrame e salvar

```python
df = pd.DataFrame({
    "tipo_gerador":              tipo_gerador,
    "tipo_atividade":            tipo_atividade,
    "bairro":                    bairro,
    "tipo_residuo_principal":    tipo_residuo_principal,
    "residuo_organico_principal": residuo_organico_principal,
    "origem_residuo":            origem_residuo,
    "quantidade_semanal":        quantidade_semanal,
    "num_pessoas":               num_pessoas,
    "destino_atual":             destino_atual,
    "custo_destinacao":          custo_destinacao,
    "potencial_reaproveitamento": potencial_reaproveitamento,
    "caracteristica_residuo":    caracteristica_residuo,
    "tentativa_reutilizacao":    tentativa_reutilizacao,
    "interesse_aprender":        interesse_aprender
})

```

```python
print(f"Dataset gerado com {df.shape[0]} linhas e {df.shape[1]} colunas.")
df.head()
```

```python
df.to_csv("../dados/simulados/respostas_simuladas.csv", index=False)
```
